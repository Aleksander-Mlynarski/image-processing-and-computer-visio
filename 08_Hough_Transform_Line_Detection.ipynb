### Hough Transform in the ab Parameter Space ‚Äî Homework

The œÅ‚ÄìŒ∏ space is not the only parameterization where each point votes for a line. An alternative is the slope‚Äìintercept form:

$y = ax + b$

In this space, a bundle of lines through a common point maps to a single point.

**Task:** Implement a function that takes a binary image and parameters:

- `aMin` ‚Äî minimum value of parameter *a*
- `aMax` ‚Äî maximum value of parameter *a*
- `aSkok` ‚Äî step size for parameter *a*
- `bMin` ‚Äî minimum value of parameter *b*
- `bMax` ‚Äî maximum value of parameter *b*
- `bSkok` ‚Äî step size for parameter *b*

The function should return the ab Hough accumulator matrix `H`.

**Notes:**

- The problem looks involved but mainly requires careful indexing of `H`.
- Build vectors `A` and `B` of all admissible *a* and *b* values (use `np.arange` over the given ranges and steps). Their indices become coordinates in `H`.
- Allocate `H` with `np.zeros`, sized from `A.shape` and `B.shape`.
- For each edge pixel in the image:
  - For every value in `A`, compute the corresponding *b*; `a_idx` is one coordinate in `H`, and `b_idx = np.argmin(abs(B - b))` gives the other.
  - Increment `H[a_idx, b_idx]`.
- Test on points generated in the "Hough Transform for a Small Point Set" editor. Suggested parameters:
  - `aMin`: -5
  - `aMax`: 5
  - `aSkok`: 0.05
  - `bMin`: -100
  - `bMax`: 100
  - `bSkok`: 1

Consider for which lines the ab representation fails.

```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
from skimage.transform import hough_line, hough_line_peaks

def hab(img, amin, amax, ask, bmin, bmax, bsk):
    A = np.arange(amin, amax, ask)
    B = np.arange(bmin, bmax, bsk)
    H = np.zeros((len(A), len(B)), dtype=np.uint8)
    
    ys, xs = np.nonzero(img)
    
    for x, y in zip(xs, ys):
        for i, a in enumerate(A):
            b = y - a * x
            j = np.argmin(np.abs(B - b))
            H[i, j] += 1
            
    return H

im = np.zeros((64, 64), dtype=np.uint8)
im[18, 31] = 1
im[40, 50] = 1

H = hab(im, -5, 5, 0.05, -100, 100, 1)

plt.imshow(H, cmap='gray', aspect='auto')
plt.show()
```

The ab representation fails for vertical lines (where $x = \mathrm{const}$), because the slope $a$ approaches infinity and would require an unbounded parameter space.
